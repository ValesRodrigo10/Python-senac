```python
# Códigos ANSI para cores
RED = '\033[91m'
GREEN = '\033[92m'
YELLOW = '\033[93m'
BLUE = '\033[94m'
MAGENTA = '\033[95m'
CYAN = '\033[96m'
BOLD = '\033[1m'
RESET = '\033[0m'

reservas = {
    100: {'nome': 'fulano', 'dias': 10, 'valor_da_diaria': 10}
}

def novaReserva(quarto, nome, dias, valorDiaria):
    if quarto in reservas:
        print(f"{RED}O quarto {quarto} já está reservado.{RESET}")
        return
    reservas[quarto] = {'nome': nome, 'dias': dias, 'valor_da_diaria': valorDiaria}
    total = dias * valorDiaria
    print(f"\n{GREEN}{BOLD}Reserva efetuada com sucesso!{RESET}")
    print(f"{CYAN}Hóspede: {nome}")
    print(f"Quarto: {quarto}")
    print(f"Dias reservados: {dias}")
    print(f"Valor da diária: R${valorDiaria:.2f}")
    print(f"Valor total da estadia: R${total:.2f}{RESET}")

def todasReservas():
    print(f"{BOLD}{MAGENTA}Todos os hóspedes cadastrados:{RESET}")
    for quarto, info in reservas.items():
        total_estadia = info['dias'] * info['valor_da_diaria']
        print(f"{YELLOW}Nome: {info['nome']}, Quarto: {quarto}, Dias: {info['dias']}, "
              f"Diária: R${info['valor_da_diaria']:.2f}, Total: R${total_estadia:.2f}{RESET}")

def buscarReservas(numero_quarto):
    if numero_quarto in reservas:
        info = reservas[numero_quarto]
        total_estadia = info['dias'] * info['valor_da_diaria']
        print(f"\n{BOLD}{BLUE}Detalhes da reserva:{RESET}")
        print(f"{CYAN}{'-=' * 30}")
        print(f"Quarto: {numero_quarto}")
        print(f"Hóspede: {info['nome']}")
        print(f"Dias: {info['dias']}")
        print(f"Valor da diária: R${info['valor_da_diaria']:.2f}")
        print(f"Total da estadia: R${total_estadia:.2f}")
        print(f"{'-=' * 30}{RESET}")
    else:
        print(f"{RED}Nenhuma reserva encontrada para o quarto {numero_quarto}.{RESET}")

def atualizarDias(quarto, nova_quantidade):
    if quarto in reservas:
        reservas[quarto]['dias'] = nova_quantidade
        print(f"{GREEN}Quantidade de dias para o quarto {quarto} atualizada para {nova_quantidade} dias.{RESET}")
    else:
        print(f"{RED}Quarto {quarto} não encontrado nas reservas.{RESET}")

def removerReserva(quarto):
    if quarto in reservas:
        nome = reservas[quarto]["nome"]
        del reservas[quarto]
        print(f"{GREEN}Reserva para o quarto {quarto}, de {nome}, foi removida com sucesso!{RESET}")
    else:
        print(f"{RED}Quarto não encontrado.{RESET}")

# Loop principal
while True:
    print(f'\n{BOLD}{CYAN}{"="*40}')
    print('Maré Alta Hotel'.center(40))
    print('O que deseja fazer hoje?'.center(40))
    print(f'{"="*40}{RESET}')
    print(f'''{YELLOW}[ 1 ]{RESET} Cadastrar um novo hóspede
{YELLOW}[ 2 ]{RESET} Exibir todas as reservas
{YELLOW}[ 3 ]{RESET} Buscar reservas
{YELLOW}[ 4 ]{RESET} Atualizar dias de uma reserva
{YELLOW}[ 5 ]{RESET} Cancelar uma reserva
{YELLOW}[ 6 ]{RESET} Sair''')

    try:
        escolha = int(input(f'\nEscolha uma opção:'))
    except ValueError:
        print(f"{RED}Por favor, digite um número válido.{RESET}")
        continue

    if escolha == 6:
        break

    match escolha:
        case 1:
            try:
                quarto = int(input('Número do quarto: '))
                nome = input('Nome do hóspede: ')
                dias = int(input('Dias: '))
                preco = float(input('Valor da diária: R$ '))
                novaReserva(quarto, nome, dias, preco)
            except ValueError:
                print(f"{RED}Erro: digite os valores corretamente.{RESET}")
        case 2:
            print('-=' * 55)
            todasReservas()
            print('-=' * 55)
        case 3:
            try:
                quarto = int(input('Digite o número do quarto que deseja buscar: '))
                buscarReservas(quarto)
            except ValueError:
                print(f"{RED}Número de quarto inválido.{RESET}")
        case 4:
            try:
                quarto = int(input('Digite o número do quarto: '))
                qtd = int(input('Digite a nova quantidade de dias: '))
                atualizarDias(quarto, qtd)
            except ValueError:
                print(f"{RED}Entrada inválida.{RESET}")
        case 5:
            try:
                quarto = int(input('Digite o número do quarto: '))
                removerReserva(quarto)
            except ValueError:
                print(f"{RED}Número de quarto inválido.{RESET}")
        case _:
            print(f"{RED}Opção desconhecida.{RESET}")

    sair = input(f'\nDeseja continuar? [S / N]:')
    if sair.lower() != 's':
        break

print(f'{BOLD}{MAGENTA}{"-=" * 30}')
print('Maré Alta Hotel ENCERRADO'.center(60))
print(f'{"-=" * 30}{RESET}')
```
